### extract random 100 complexes from val set

```python
import csv
import random
```

```python
val_lines = []

with open('/data/rsg/nlp/sdobers/data/DIPS/data_file_only_val.csv') as file_obj:
    heading = next(file_obj)
    reader_obj = csv.reader(file_obj)
    for row in reader_obj:
        val_lines.append(row[0])
val_lines
```

```python
random_subset = random.sample(val_lines, k=100)
random_subset
```

```python
random_subset_to_save = [name + ',val\n' for name in random_subset]
random_subset_to_save.insert(0,'path,split\n')
random_subset_to_save
```

```python
with open("/data/rsg/nlp/sdobers/data/DIPS/data_file_val_random100.csv", "w") as fp:
    fp.writelines(random_subset_to_save)
```

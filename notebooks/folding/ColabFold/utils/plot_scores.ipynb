# Using scores.json for plotting

Our example is a complex where you can clearly see the chain break after 93 residues

```python
from pathlib import Path
import numpy
import json
import matplotlib.pyplot as plt
```

```python
scores = json.loads(Path("3G5O_A_3G5O_B_unrelaxed_rank_1_model_1_scores.json").read_text())
```

```python
plt.imshow(numpy.asarray(scores["pae"]))
plt.xlabel("Residue 1")
plt.ylabel("Residue 2")
plt.show()
```

```python
plt.plot(numpy.asarray(scores["plddt"]))
plt.xlabel("Residue")
plt.ylabel("average pLDDT")
plt.show()
```

```python
print("pTMscore:", scores["ptm"])
```
